### 19. Explique la codificacin Base64.

Base 64 es un sistema de numeracin posicional que usa 64 como base. Es la mayor potencia de dos que puede ser representada usando nicamente los caracteres imprimibles de ASCII. Todas las variantes famosas que se conocen con el nombre de Base64 usan el rango de caracteres A-Z, a-z y 0-9 en este orden para los primeros 62 dgitos, pero los smbolos escogidos para los ltimos dos dgitos varan considerablemente de unas a otras.

Base 64 se codifica tomando los bytes de tres como entrada en cuatro sextetos, o sea de a 24 bits. Cuando esta cantidad no coincide, completamos con el carcter de relleno = los sextetos restantes.

Base64 posee una caracterstica interesante: tiene la capacidad de convertir cualquier dato definido a nivel de bytes en un formato seguro de transportar por Internet como lo son los caracteres ASCII. Esto significa que cualquier archivo ya sea, de texto plano o archivo ejecutable, archivo binario, etc. puede transportarse por Internet sin perder su esencia.

Funciona de la siguiente manera, la codificacin Base64 toma los datos binarios originales y los opera dividindolos en tokens de tres bytes. Un byte consta de ocho bits, por lo que Base64 toma 24 bits en total. Estos 3 bytes se convierten luego en cuatro caracteres imprimibles del estndar ASCII.

El primer paso es tomar los tres bytes (24 bits) de datos binarios y dividirlos en cuatro nmeros de seis bits. Debido a que el estndar ASCII define el uso de siete bits, Base64 solo usa 6 bits (que corresponden a 2 ^ 6 = 64 caracteres) para garantizar que los datos codificados sean imprimibles y que no se use ninguno de los caracteres especiales disponibles en ASCII.

La conversin ASCII de grupos de 3 bytes y 24 bits se repite hasta que se codifica toda la secuencia de bytes de datos originales.

# <u>Parte 2: Programacin Cliente / Servidor utilizando la interfaz sockets.</u>
Servidor:

Decidimos crear el script en c, lo guardamos en un archivo llamado servidor.c.


```python
%%writefile servidor.c

#include <sys/types.h>
#include <sys/socket.h>
#include <netinet/in.h>
#include <string.h>
#include <stdlib.h>
#include <stdio.h>
#define PORTNUMBER 12345

int main (int argc, char **argv){
    char buf[10];
    int s, n, ns, len;
    struct sockaddr_in direcc;

    s = socket(AF_INET,SOCK_STREAM, 0);
    
    bzero((char*)&direcc, sizeof(direcc));
    direcc.sin_family = AF_INET;
    direcc.sin_port = htons(PORTNUMBER);
    direcc.sin_addr.s_addr = htonl(INADDR_ANY);

    len = sizeof(struct sockaddr_in);
    bind(s, (struct sockaddr*)&direcc, len);
    listen(s, 5);
    ns = accept(s, (struct sockaddr*)&direcc, &len);
    
    while ((n = recv(ns, buf, sizeof(buf), 0)) > 0)
    {   write(1, buf,n);
        
    }
    close(ns);close(s);
    exit(0);
}
```

### 2. Modifique el servidor para que no finalice.

Para que el servidor no finalice, agregamos un while(1). Este while(1), encierra la parte del servidor que se encarga de crear un nuevo socket y guardar en un buffer los mensajes del cliente.  


```python
%%writefile servidor2.c

#include <sys/types.h>
#include <sys/socket.h>
#include <netinet/in.h>
#include <string.h>
#include <stdlib.h>
#include <stdio.h>

#define PORTNUMBER  12345

int main(void){
   char buf[10];
   int s, n, ns, len;
   struct sockaddr_in direcc;
  
   s = socket(AF_INET, SOCK_STREAM, 0);

   bzero((char *) &direcc, sizeof(direcc));
   direcc.sin_family = AF_INET;
   direcc.sin_port = htons(PORTNUMBER);
   direcc.sin_addr.s_addr = htonl(INADDR_ANY);

   len = sizeof(struct sockaddr_in);
   bind(s, (struct sockaddr *) &direcc, len);
   listen(s, 5);
   while(1){
	   ns = accept(s, (struct sockaddr *) &direcc, &len);

	   while ((n = recv(ns, buf, sizeof(buf), 0)) > 0)
	       write(1, buf, n);
    }    
   close(ns); close(s);
   exit(0);
}
```

```python
%%bash
gcc -o servidor2 servidor2.c
./servidor2
```
