### 19. Explique la codificación Base64.

Base 64 es un sistema de numeración posicional que usa 64 como base. Es la mayor potencia de dos que puede ser representada usando únicamente los caracteres imprimibles de ASCII. Todas las variantes famosas que se conocen con el nombre de Base64 usan el rango de caracteres A-Z, a-z y 0-9 en este orden para los primeros 62 dígitos, pero los símbolos escogidos para los últimos dos dígitos varían considerablemente de unas a otras.

Base 64 se codifica tomando los bytes de tres como entrada en cuatro sextetos, o sea de a 24 bits. Cuando esta cantidad no coincide, completamos con el carácter de relleno “=” los sextetos restantes.

Base64 posee una característica interesante: tiene la capacidad de convertir cualquier dato definido a nivel de bytes en un formato seguro de transportar por Internet como lo son los caracteres ASCII. Esto significa que cualquier archivo ya sea, de texto plano o archivo ejecutable, archivo binario, etc. puede transportarse por Internet sin perder su “esencia”.

Funciona de la siguiente manera, la codificación Base64 toma los datos binarios originales y los opera dividiéndolos en tokens de tres bytes. Un byte consta de ocho bits, por lo que Base64 toma 24 bits en total. Estos 3 bytes se convierten luego en cuatro caracteres imprimibles del estándar ASCII.

El primer paso es tomar los tres bytes (24 bits) de datos binarios y dividirlos en cuatro números de seis bits. Debido a que el estándar ASCII define el uso de siete bits, Base64 solo usa 6 bits (que corresponden a 2 ^ 6 = 64 caracteres) para garantizar que los datos codificados sean imprimibles y que no se use ninguno de los caracteres especiales disponibles en ASCII.

La conversión ASCII de grupos de 3 bytes y 24 bits se repite hasta que se codifica toda la secuencia de bytes de datos originales.

# <u>Parte 2: Programación Cliente / Servidor utilizando la interfaz sockets.</u>

Decidimos crear el script en c, lo guardamos en un archivo llamado “servidor.c”.


````python
%%writefile servidor.c

#include <sys/types.h>
#include <sys/socket.h>
#include <netinet/in.h>
#include <string.h>
#include <stdlib.h>
#include <stdio.h>
#define PORTNUMBER 12345

int main (int argc, char **argv){
    char buf[10];
    int s, n, ns, len;
    struct sockaddr_in direcc;

    s = socket(AF_INET,SOCK_STREAM, 0);
    
    bzero((char*)&direcc, sizeof(direcc));
    direcc.sin_family = AF_INET;
    direcc.sin_port = htons(PORTNUMBER);
    direcc.sin_addr.s_addr = htonl(INADDR_ANY);

    len = sizeof(struct sockaddr_in);
    bind(s, (struct sockaddr*)&direcc, len);
    listen(s, 5);
    ns = accept(s, (struct sockaddr*)&direcc, &len);
    
    while ((n = recv(ns, buf, sizeof(buf), 0)) > 0)
    {   write(1, buf,n);
        
    }
    close(ns);close(s);
    exit(0);
}
````

### 3. Modifique el servidor para que el servicio que presta se implemente en una función separada del cuerpo principal del programa (main()).

Para separar el servicio que presta el servidor del cuerpo del programa principal lo que hicimos fue crear una función llamada funcion_servidor que recibe un entero que representa el nuevo socket, un puntero a char que representa el buffer, y un entero que representa el número de bytes leídos en el búfer.  

````python
%%writefile servidor3.c

#include <sys/types.h>
#include <sys/socket.h>
#include <netinet/in.h>
#include <string.h>
#include <stdlib.h>
#include <stdio.h>

#define PORTNUMBER  12345


void funcion_servidor(int ns, char * buf, int n){
  while ((n = recv(ns, buf, sizeof(buf), 0)) > 0)
    write(1,buf,n);
}

int main(void){
   char buf[10];
   int s, n, ns, len;
   struct sockaddr_in direcc;
  
   s = socket(AF_INET, SOCK_STREAM, 0);

   bzero((char *) &direcc, sizeof(direcc));
   direcc.sin_family = AF_INET;
   direcc.sin_port = htons(PORTNUMBER);
   direcc.sin_addr.s_addr = htonl(INADDR_ANY);

   len = sizeof(struct sockaddr_in);
   bind(s, (struct sockaddr *) &direcc, len);
   listen(s, 5);
   while(1){
     ns = accept(s, (struct sockaddr *) &direcc, &len);

     funcion_servidor(ns,buf,n);
    }    
   close(ns); close(s);
   exit(0);
}
````

### 4. Modifique el servidor para que preste el servicio de echo (vea RFC862).



El protocolo Echo es un servicio de la Familia de protocolos de Internet definido en el RFC 862. Fue propuesto originalmente para permitir la comprobación de la red y la medición del tiempo de ida y vuelta en las redes IP.  
El servidor devuelve una copia idéntica a los datos que recibió.

Para realizar esto lo que hicimos fue utilizar la función de la librería “socket.h” llamada “send” la cual tiene la siguiente sintaxis:

```
send(int sockfd, const void *buf, size_t len, int flags);
```
Se creó un nuevo archivo llamado servidor2.c el cual quedó conformado del siguiente modo:


````python
%%writefile servidor4.c

#include <sys/types.h>
#include <sys/socket.h>
#include <netinet/in.h>
#include <string.h>
#include <stdlib.h>
#include <stdio.h>
#define PORTNUMBER 12345

int main (void){
    char buf[10];
    int s, n, ns, len;
    struct sockaddr_in direcc;

    s = socket(AF_INET,SOCK_STREAM, 0);
    
    bzero((char*)&direcc, sizeof(direcc));
    direcc.sin_family = AF_INET;
    direcc.sin_port = htons(PORTNUMBER);
    direcc.sin_addr.s_addr = htonl(INADDR_ANY);

    len = sizeof(struct sockaddr_in);
    bind(s, (struct sockaddr*)&direcc, len);
    listen(s, 5);
    ns = accept(s, (struct sockaddr*)&direcc, &len);
    
    while ((n = recv(ns, buf, sizeof(buf), 0)) > 0)
    {
        write(1, buf, n);
        send(ns, buf, n, 0);
    }
    close(ns);close(s);
    exit(0);
}
````

````python
%%bash 
gcc -o servidor4 servidor4.c
./servidor4
````

Como se puede observar, se utilizó la función send en el cuerpo del “While”.

Al escribir un mensaje en el cliente y enviarlo ( presionar “Enter”) se imprime este mensaje en la salida estándar del servidor, pero también lo hace en la salida estándar del cliente, es esto el funcionamiento del protocolo echo, como se buscaba que funcione. También se puede observar como al cerrar la conexión en el cliente se cierra el servidor.

````python
%%writefile servidor5.c

#include <sys/types.h>
#include <sys/socket.h>
#include <netinet/in.h>
#include <string.h>
#include <stdlib.h>
#include <stdio.h>
#define PORTNUMBER 12345

int main (int argc, char **argv){
    
    char buf[100];
    int s, n, ns, len;
    struct sockaddr_in direcc;

    s = socket(AF_INET,SOCK_STREAM, 0);
    
    bzero((char*)&direcc, sizeof(direcc));
    direcc.sin_family = AF_INET;
    direcc.sin_port = htons(PORTNUMBER);
    direcc.sin_addr.s_addr = htonl(INADDR_ANY);

    len = sizeof(struct sockaddr_in);
    bind(s, (struct sockaddr*)&direcc, len);
    listen(s, 5);
    ns = accept(s, (struct sockaddr*)&direcc, &len);
    
    while ((n = recv(ns, buf, sizeof(buf), 0)) > 0){
        int cantidadCaracteres = 0;
        char buffCaracter[4];
        write(1, buf, n);
        while(cantidadCaracteres < 10){
            sprintf(buffCaracter,"%d\n",cantidadCaracteres);
            sleep(1);
            send(ns, buffCaracter, sizeof(buffCaracter), 0);
            cantidadCaracteres +=1;
        }
    }
    close(ns);close(s);
    exit(0);  
   
}
````

### 8. Transformar el Servidor iterativo TCP de Eco del punto 4), en un servidor iterativo de Eco con UDP. Crear el cliente correspondiente.

Los sockets UDP son sockets no orientados a conexión. Esto quiere decir que un programa puede abrir un socket y ponerse a escribir mensajes en él o leer, sin necesidad de esperar a que alguien se conecte en el otro extremo del socket.
El protocolo UDP, al no ser orientado a conexión, no garantiza que el mensaje llegue a su destino. Parece claro que si mi programa envía un mensaje y no hay nadie escuchando, ese mensaje se pierde. De todas formas, aunque haya alguien escuchando, el protocolo tampoco garantiza que el mensaje llegue. Lo único que garantiza es, que si llega, llega sin errores.



El código fuente del ejecutable es el siguiente:
Con respecto a la imagen anterior, podemos ver que el cliente crea el socket con la función socket() usada en la línea 21. Se indica que va ser un cliente UDP con el parámetro SOCK_DGRAM.
Utilizamos la función memset() para limpiar la variable usada para recibir la información del servidor.
En la imagen se puede observar que se realiza un ciclo infinito en el que se limpia la salida estándar, se lee la entrada estándar y se verifica si el usuario desea continuar enviando datos o salir. Si desea salir, se termina la sesión del cliente y se cierra, si desea continuar se envía el mensaje recién ingresado al servidor. Luego se mantiene en espera de recibir un mensaje, al recibirlo lo imprime en la salida estándar y se reinicia el ciclo.

````python
%%writefile servidor8.c
#include <sys/types.h>
#include <sys/socket.h>
#include <netinet/in.h>
#include <arpa/inet.h>
#include <stdio.h>
#include <unistd.h>
#include <errno.h>
#include <string.h>
#include <stdlib.h>
#define BUFLEN 256
#define PORTNUMBER 8888
int main()
{
    int sock;
    int addr_len, bytes_read, bytes_write;
    char recv_data[BUFLEN];
    struct sockaddr_in server_addr , client_addr;
//Creamos el socket al que nos vamos a comunicar.
    if ((sock = socket(AF_INET, SOCK_DGRAM, 0)) == -1) {
        perror("Socket");
        exit(1);
    }
    server_addr.sin_family = AF_INET;
    server_addr.sin_port = htons(PORTNUMBER);
    server_addr.sin_addr.s_addr = INADDR_ANY;
    bzero(&(server_addr.sin_zero),8);
    if (bind(sock,(struct sockaddr *)&server_addr, sizeof(struct sockaddr)) == -1)
    {
        perror("Bind");
        exit(1);
    }
    addr_len = sizeof(struct sockaddr);
    printf("UDPServer Waiting for client on port %d\n",PORTNUMBER);
    fflush(stdout);
    while (1)
    {
        printf("Esperando información...\n");
        fflush(stdout);
        if ((bytes_read = recvfrom(sock,recv_data, BUFLEN, 0, (struct sockaddr *) &client_addr, &addr_len)) == -1)
        {
            printf("ERROR RECIBIENDO MENSAJE\n");
            break;
        }
        recv_data[bytes_read] = '\0';
        //Devolvemos lo recibido (hacemos el ECHO)
        if (bytes_write = sendto(sock,recv_data,strlen(recv_data),0,(struct sockaddr *)&client_addr,sizeof(struct sockaddr)) == -1)
        {
            printf("ERROR ENVIANDO MENSAJE.\n");
            break;
        }
        printf("se recibió desde el servidor: %s\n",recv_data);
    }
    return 0;
}
````
